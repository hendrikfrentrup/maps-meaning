```python
import pyspark
from operator import add

sc = pyspark.SparkContext()
```

```python
textFile = sc.textFile("data/shakespeare.txt").filter(lambda l: len(l) > 0)
words=textFile.flatMap(lambda l: l.split())

counts = words.map(lambda w: (w, 1)).reduceByKey(lambda x,y: x+y).map(lambda x: (x[1],x[0])).sortByKey(False) #(add)
```

```python
counts.collect()
```
